```python
import random as rd
from random import randrange
```

Se simularan ambos eventos de 10 en 10 horas, cada cantidad de horas con su respectiva probabilidad de fallar y sus costos

```python
#numero de eventos
n=10000
#contador de eventos
c=1
#lista de horas
h=[2,3,4,5,6,7,8,9]
#contados de la lista
i=0
#numero de averias
av=0
#horas de trabajo
hrs=0
#primer while sirve para contar numero de eventos, cada que sucede una falla se contabiliza como un evento
while c<=n:
#segundo while avanza cada 10 horas y analiza la posibilidad de alguna averia
  while 7>i and c<=n:
    p=random.randrange(100)
    if h[i]==2:
      if p<=5:
        av=av+1
        i=-1
        c=c+1
        hrs=hrs+h[i]
    if h[i]==3 or h[i]==4 or h[i]==7:
      if p<=15:
        av=av+1
        i=-1
        c=c+1
        hrs=hrs+h[i]
    if h[i]==5 or h[i]==6:
      if p<=20:
        av=av+1
        i=-1
        c=c+1
        hrs=hrs+h[i]
    if h[i]==8:
      if p<=10:
        av=av+1
        i=-1
        c=c+1
        hrs=hrs+h[i]
      else:
        c=c+1
        hrs=hrs+h[i]
    i=i+1
  i=0
  c=c+1
#costo para el primer caso
cost=av*110
#promedio de los costo por cantidad de horas
prom=cost/hrs
print("De",hrs,"horas de trabajo se tiene un costo de",cost,"dando asi",prom,"euros por hora trabajada")
c=1
i=1
av=0
hrs=0
while c<=n:
  while 8>i and c<=n:
    p=random.randrange(100)
    if h[i]==3:
      if p<=5:
        av=av+1
        i=-1
        c=c+1
        hrs=hrs+h[i]
    if h[i]==4 or h[i]==5 or h[i]==8:
      if p<=15:
        av=av+1
        i=-1
        c=c+1
        hrs=hrs+h[i]
    if h[i]==6 or h[i]==7:
      if p<=20:
        av=av+1
        i=-1
        c=c+1
        hrs=hrs+h[i]
    if h[i]==9:
      if p<=10:
        av=av+1
        i=-1
        c=c+1
        hrs=hrs+h[i]
      else:
        c=c+1
        hrs=hrs+h[i]
    i=i+1
  i=0
  c=c+1
#costo para el primer caso
cost=av*250
#promedio de los costo por cantidad de horas
prom=cost/hrs
print("De",hrs,"horas de trabajo se tiene un costo de",cost,"dando asi",prom,"euros por hora trabajada")
```
